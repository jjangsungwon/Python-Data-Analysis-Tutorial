## KNN Algorithm

K-Nearest Neighbor Algorithm (근접 이웃 알고리즘)
- KNN은 지도학습(Supervised Learning) 알고리즘입니다.
- 다양한 레이블의 데이터 중에서, 자신과 가까운 데이터를 찾아 자신의 레이블을 결정하는 방식입니다.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 각 데이터의 위치: 25 X 2 크기에 각각 0 ~ 100
trainData = np.random.randint(0, 100, (25, 2)).astype(np.float32)
# 각 데이터는 0 or 1
response = np.random.randint(0, 2, (25, 1)).astype(np.float32)

# 값이 0인 데이터를 각각 (x, y) 위치에 빨간색으로 칠합니다.
red = trainData[response.ravel() == 0]
plt.scatter(red[:, 0], red[:, 1], 80, 'r', 's')  # 빨간색 네모
# 값이 1인 데이터를 각각 (x, y) 위치에 파란색으로 칠합니다.
blue = trainData[response.ravel() == 1]
plt.scatter(blue[:, 0], blue[:, 1], 80, 'b', 'o')  # 파란색 동그라미

# (0 ~ 100, 0 ~ 100) 위치의 데이터를 하나 생성합니다.
new_data = np.random.randint(0, 100, (1, 2)).astype(np.float32)
plt.scatter(new_data[:, 0], new_data[:, 1], 80, 'g', '^')  # 초록색 세모

knn = cv2.ml.KNearest_create()
# ROW_SAMPLE 인자는 데이터 하나가 한 행으로 들어가는 것을 의미합니다.
knn.train(trainData, cv2.ml.ROW_SAMPLE, response)
ret, results, neighbours, dist = knn.findNearest(new_data, 3)

# 가까운 3개의 레이블을 찾고, 거리를 고려하여 자신의 레이블을 정합니다.
print("result :", results)
print("neighbours :", neighbours)
print("distance: ", dist)

plt.show()
```

# OpenCV 기본 사용법

## OpenCV란?
- 영상 처리와 컴퓨터 비전을 위한 오픈소스 라이브러리
- C, C++, Python 등에서 사용 가능

## OpenCV의 기본적인 함수

### cv2.imread(file_name, flag)
이미지를 읽어 Numpy 객체로 만드는 함수입니다.
- file_name: 읽고자 하는 이미지 파일  
- flag: 이미지를 읽는 방법 설정  

IMREAD_COLOR: 이미지를 COLOR로 읽고, 투명한 부분은 무시  
IMREAD_GRAYSCALE: 이미지를 Graysacle로 읽기  
IMREAD_UNCHANGED: 이미지를 Color로 읽고, 투명한 부분도 읽기(Alpha)  

반환 값: Numpy 객체 (행, 열, 색상: 기본 BGR)

### cv2.imshow(title, image)
특정한 이미지를 화면에 출력하는 함수입니다.
- title: 윈도우 창의 제목
- image: 출력할 이미지 객체

### cv2.imwrite(file_name, image)
특정한 이미지를 파일로 저장하는 함수입니다.
- file_name: 저장할 이미지 파일 이름
- image: 저장할 이미지 객체

### cv2.waitKey(time)
키보드 입력을 처리하는 함수입니다.
- time: 입력 대기 시간 (무한대기: 0)

반환 값: 사용자가 입력한 Ascii Code

### cv2.destroyAllWindows()
화면의 모든 윈도우를 닫는 함수입니다.

### cv2.cvtColor(images, flag)
- 이미지의 색상 형태를 변경하는 함수입니다.

## 이미지 출력
- OpenCV는 BGR을 기준으로 하며, Matplotlib는 RGB를 기준으로 합니다.
- cv2.cvtColor 함수를 사용하면 이미지의 색상 형태를 변경할 수 있습니다.

```python
import cv2
import matplotlib.pyplot as plt

img_basic = cv2.imread('../images/cat.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))
plt.show()

img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))
plt.show()
```
